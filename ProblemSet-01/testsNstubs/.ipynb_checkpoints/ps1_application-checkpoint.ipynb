```python
def usps():
    ''' performs the usps analysis for assignment 4'''
    images, labels = imp.load_data('usps')
    x, y, n = images.shape
    total_class = labels.shape[0]
    n_rows = 10
    n_cols = 3
    k = 25 # Total number of the largest principal values that should be depicted
    
    # Visualizing pca from the original data
    title = 'Visualizing PCA from the original images'
    fig_pca, axes_pca = pl.subplots(nrows=n_rows, ncols=n_cols, figsize=(5 * n_cols, 2.4 * n_rows))
    fig_pca.suptitle(title, fontweight='bold', color='red', fontsize=14)
    
    for i in range(total_class):
        cls_idx = np.where(labels[i, :] == 1)[0] # Get indices of images that belong to particular class
        cls_images2D, fact = imp.im2vec(images[:, :, cls_idx]) # Convert 3D images to 2D images
        Z, U, D = imp.pca(cls_images2D, x*y) # Calculate PCA for a given class
        
        visualize_pca(Z, U, D, (axes_pca[i, 0], axes_pca[i, 1], axes_pca[i, 2]), i, 25) # Visualize PCA
    
    # Visualizing pca from noisy data
    denoised_images = [] # List to store 3 various denoised images (moderate, very, extrem)
    noisy_images = []
    for idx, sigma in enumerate([moderately_noisy, very_noisy, extrem_noisy]):
        noisy_img = get_noisy_images(images, sigma)
        noisy_images.append(noisy_img)
        
        title = get_noisy_title(sigma)
        fig_pca_noisy, axes_pca_noisy = pl.subplots(nrows=n_rows, ncols=n_cols, figsize=(5 * n_cols, 2.4 * n_rows))
        fig_pca_noisy.suptitle(title, fontweight='bold', color='red', fontsize=14)
        
        Y = np.zeros([x, y, n]) # Denoised images
        for i in range(total_class):
            cls_idx = np.where(labels[i, :] == 1)[0] # Get indices of images that belong to particular class
            cls_images2D, fact = imp.im2vec(noisy_img[:, :, cls_idx]) # Convert 3D images to 2D images
            Z, U, D = imp.pca(cls_images2D, x*y) # Calculate PCA for a given class
            
            visualize_pca(Z, U, D, (axes_pca_noisy[i, 0], axes_pca_noisy[i, 1], axes_pca_noisy[i, 2]), i, 25) # Visualize PCA

            imgs, fact = imp.im2vec(images[:, :, cls_idx])
            im = np.mean(imgs, axis=1)[:, np.newaxis] + np.dot(U, Z) # Denoise images
            Y[:, :, cls_idx] = imp.vec2im(im)
            
        denoised_images.append(Y)
        
        
        
    # Visualize 10 examples of original images, noisy images, and denoised reconstruction
    Y_moderate = denoised_images[0]
    Y_very = denoised_images[1]
    Y_extrem = denoised_images[2]
    
    X_moderate = noisy_images[0]
    X_very = noisy_images[1]
    X_extrem = noisy_images[2]
    
    idx = [3, 45, 78, 105, 140, 211, 399, 480, 520, 645]
    title = 'Visualize 10 examples of original images, noisy images, and denoised reconstruction'
    #fig_ex, axes_ex = pl.subplots(nrows = 5, ncols = 2, figsize=(10, 8))
    
    fig_ex = pl.figure(figsize=(15, 12))
    fig_ex.suptitle(title, fontweight='bold', color='red', fontsize=14)
    gs = gridspec.GridSpec(4, 2)
    ax1 = fig_ex.add_subplot(gs[0, :])
    ax2 = fig_ex.add_subplot(gs[1, 0])
    ax3 = fig_ex.add_subplot(gs[1, 1])
    ax4 = fig_ex.add_subplot(gs[2, 0])
    ax5 = fig_ex.add_subplot(gs[2, 1])
    ax6 = fig_ex.add_subplot(gs[3, 0])
    ax7 = fig_ex.add_subplot(gs[3, 1])

    imp.visualize_data(ax1, images[:, :, idx], (1,10))
    ax1.set_title('Original images', fontweight='bold', fontsize=12)
    
    imp.visualize_data(ax2, X_moderate[:, :, idx], (1,10))
    ax2.set_title('Moderately noisy images', fontweight='bold', fontsize=12)
    
    imp.visualize_data(ax3, Y_moderate[:, :, idx], (1,10))
    ax3.set_title('Reconstruction moderately noisy images', fontweight='bold', fontsize=12)
    
    imp.visualize_data(ax4, X_very[:, :, idx], (1,10))
    ax4.set_title('Very noisy images', fontweight='bold', fontsize=12)
    
    imp.visualize_data(ax5, Y_very[:, :, idx], (1,10))
    ax5.set_title('Reconstruction very noisy images', fontweight='bold', fontsize=12)
    
    imp.visualize_data(ax6, X_extrem[:, :, idx], (1,10))
    ax6.set_title('Extremely noisy images', fontweight='bold', fontsize=12)
    
    imp.visualize_data(ax7, Y_extrem[:, :, idx], (1,10))
    ax7.set_title('Reconstruction extremely noisy images', fontweight='bold', fontsize=12)
    
```

```python
def visualize_pca(Z, U, D, axes, cls, k):
    '''
        Z = m x n Matrix
        U = d x d Matrix
        D = 1 x d Matrix
        axes = 
        cls =
        k = 
    '''
    m, n = Z.shape
      
    axes[0].bar(np.arange(len(D)), D)
    axes[0].set_title('All principle values of class ' + str(cls), fontweight='bold', fontsize=10)
        
    axes[1].bar(np.arange(k), D[:k])
    axes[1].set_title('The largest 25 principle values of class ' + str(cls), fontweight='bold', fontsize=10)
    
    U_3D = imp.vec2im(U[:, :5], (16, 16))
    imp.visualize_data(axes[2], U_3D, (1,5))
    axes[2].set_title('The first 5 principle directions of class ' + str(cls), fontweight='bold', fontsize=10)
    
    pl.tight_layout(pad=5, h_pad=1.5)
```

```python
def get_noisy_images(images, sigma):
    ''' add some noise to images with standard deviation sigma'''
    x, y, n = images.shape
    return images + (sigma * np.random.randn(x, y, n))

def get_noisy_title(sigma):
    ''' get title of the figure depending on the level of the noise'''
    if sigma == moderately_noisy:
        title = 'Visualizing PCA from moderately noisy images'
    elif sigma == very_noisy:
        title = 'Visualizing PCA from very noisy images'
    elif sigma == extrem_noisy:
        title = 'Visualizing PCA from extrem noisy images'
    else:
        title = 'Visualizing PCA from noisy images with standard deviation of ' + str(sigma)
    
    return title

def denoise_images(images):
    
    return Y
```

```python
def outliers_calc():
    ''' outlier analysis for assignment 5'''
    
    d = np.load('banana_data.npz')
    data = d['data']
    label = d['label']
    #print 'data: ', data.shape
    #print 'label: ', label.shape
    pos_idx = np.nonzero(label == 1)[1]
    neg_idx = np.nonzero(label == -1)[1]
    pos_data = data[:, pos_idx]
    neg_data = data[:, neg_idx]
    #A = np.arange(5300)
    #pl.scatter(A, label)
    #pl.scatter(pos_data[0, :], pos_data[1, :], c='r')
    #pl.scatter(neg_data[0, :], neg_data[1, :], c='b')
    #print 'pos: ', pos_data.shape
    #print 'neg: ', neg_data.shape
    fig, axes = pl.subplots(nrows=3, ncols=2, figsize=(16,16))
    axes[0, 0].set_title('Positive data')
    axes[0, 0].scatter(pos_data[0, :], pos_data[1, :], c='r')
    axes[0, 1].set_title('Negative data')
    axes[0, 1].scatter(neg_data[0, :], neg_data[1, :], c='b')
    idx = 1
    idy = 0
    for rate in [0.01, 0.05, 0.1, 0.25]:
    	total_outliers = rate * pos_data.shape[1]
    	outlier_idx = np.random.choice(neg_idx, total_outliers, replace=False)
    	print 'outlier_idx: ', len(outlier_idx)
    	outlier_data = data[:, outlier_idx]
    	pos_data_outlier = np.concatenate((pos_data, outlier_data), axis=1)
    	print 'pos_data_outlier: ', pos_data_outlier.shape
        
        axes[idx, idy].scatter(pos_data[0, :], pos_data[1, :], c='r')
        axes[idx, idy].scatter(outlier_data[0, :], outlier_data[1, :], c = 'g')
        axes[idx, idy].set_title('Contamination Rate: ' + str(rate))
        idy += 1
        if idy == 2:
            idy = 0
            idx += 1
    # np.savez_compressed('outliers', var1=var1, var2=var2, ...)
```
