```python
from __future__ import division 
import numpy as np
import pylab as pl
%matplotlib inline

import ps2_implementation as imp
imp = reload(imp)

#colors = ['red', 'green', 'yellow', 'magenta', 'cyan', 'blue', 'dimgray', 'orange', 'lightblue', 'lime']
colors = ['DarkSalmon', 'green', 'yellow', 'dimgray', 'cyan', 'blue', 'magenta', 'red', 'lightblue', 'lime']
```

```python
def load_data(dataset):
    data = np.load('../ps2_datasets/' + dataset)
    #print 'data: ', data.shape
    return data
    
```

```python
data5 = load_data('5_gaussians.npy')
data2 = load_data('2_gaussians.npy')

print 'data5.shape: ', data5.shape
print 'data2.shape: ', data2.shape

fig = pl.figure(figsize=(14, 6))
ax1 = fig.add_subplot(121)
ax1.scatter(data5[0, :], data5[1, :])
ax1.set_title('5_gaussians', fontsize=18)
ax1.set_xlabel('$x_1$', fontsize=16)
ax1.set_ylabel('$x_2$', fontsize=16)

ax2 = fig.add_subplot(122)
ax2.scatter(data2[0, :], data2[1, :])
ax2.set_title('2_gaussians', fontsize=18)
ax2.set_xlabel('$x_1$', fontsize=16)
ax2.set_ylabel('$x_2$', fontsize=16)

fig.savefig('dataset.png')
```

```python
''' Assignment 6 - 5_gaussians'''
mu_list = []
r_list = []
for i in range(10):
    mu, r = imp.kmeans(data5, k=i+1)
    mu_list.append(mu)
    r_list.append(r)
    
fig, axes = pl.subplots(nrows=5, ncols=2, figsize=(5*2,5*5))
id_x, id_y = 0, 0

for i in range(10):
    for idx, cl in enumerate(np.unique(r)):
        data5_idx = data5[:, np.nonzero(r_list[i]==cl)[0]]
        axes[id_x, id_y].scatter(data5_idx[0, :], data5_idx[1, :], c=colors[idx])
        #axes[id_x, id_y].scatter(mu_list[i][0, :], mu_list[i][1, :], marker='+', color='red', s=60)
        axes[id_x, id_y].set_title('k-means with '+ str(i+1) + ' clusters')
    
    id_y += 1
    if id_y == 2:
        id_y = 0
        id_x += 1
```

```python
loss_list = np.zeros(100)
r_kmeans_list = []
mu_kmeans_list = []
for i in range(100):
    mu, r, loss = imp.kmeans(data5, k=5)
    loss_list[i] = loss
    mu_kmeans_list.append(mu)
    r_kmeans_list.append(r)
    
min_idx_kmeans = np.argmin(loss_list)
print 'min_idx_kmeans: ', min_idx_kmeans
print 'min_loss_kmeans: ', loss_list[min_idx_kmeans]
```

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.scatter(range(1,101), loss_list)
ax.set_title('K-means loss values, run 100 times', fontsize=14)
ax.set_xlabel('i-th run, $i = 1, ..., 100$', fontsize=12)
ax.set_ylabel('loss value', fontsize=12)

fig.savefig('lossvalues_kmeans.png')
```

```python
fig = pl.figure()
ax = fig.add_subplot(111)

for idx, cl in enumerate(np.unique(r_kmeans_list[min_idx_kmeans])):
    data5_idx = data5[:, np.nonzero(r_kmeans_list[min_idx_kmeans]==cl)[0]]
    ax.scatter(data5_idx[0, :], data5_idx[1, :], c=colors[idx])
    ax.scatter(mu_kmeans_list[min_idx_kmeans][0, :], mu_kmeans_list[min_idx_kmeans][1, :], marker='+', color='red', s=60)

ax.set_title('k-means with 5 clusters, loss = ' + '%.2f' % loss_list[min_idx_kmeans], fontsize=16)
ax.set_xlabel('$x_1$', fontsize=16)
ax.set_ylabel('$x_2$', fontsize=16)

fig.savefig('5gaussians_kmeans.png')
```

```python
likelihood_random_list = np.zeros(100)
sigma_random_list = []
mu_random_list = []
pi_random_list = []
for i in range(100):
    pi, mu, sigma, likelihood = imp.em_gmm(data5, k=5)
    likelihood_random_list[i] = likelihood
    mu_random_list.append(mu)
    sigma_random_list.append(sigma)
    pi_random_list.append(pi)
    
max_likelihood_random_idx = np.argmax(likelihood_random_list)
print 'max_likelihood_idx: ', max_likelihood_random_idx
print 'max_likelihood: ', likelihood_random_list[max_likelihood_random_idx]
```

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.scatter(range(1,101), likelihood_random_list)
ax.set_title('GMM log likelihood with random initialization', fontsize=14)
ax.set_xlabel('i-th run, $i = 1, ..., 100$', fontsize=12)
ax.set_ylabel('log likelihood', fontsize=12)

fig.savefig('likelihoods_gmm_random.png')
```

```python
fig = pl.figure()
ax = fig.add_subplot(111)

imp.plot_gmm_solution(data5, mu_random_list[max_likelihood_random_idx], sigma_random_list[max_likelihood_random_idx], ax)

ax.set_title('GMM with 5 clusters, log likelihood = ' + '%.2f' % likelihood_random_list[max_likelihood_random_idx], fontsize=14)
ax.set_xlabel('$x_1$', fontsize=16)
ax.set_ylabel('$x_2$', fontsize=16)

fig.savefig('5gaussians_gmm_random.png')
```
