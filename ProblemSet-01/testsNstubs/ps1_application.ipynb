```python
def usps():
    ''' Performs the usps analysis for Assignment 4'''
    
    images, labels = imp.load_data('usps') # Load usps data set
    x, y, n = images.shape # Get dimension and total data points
    total_class = labels.shape[0] # Total class of digits
    
    n_rows = 5 # Number of rows used in subplot
    n_cols = 2 # Number of cols used in subplot
    
    ''' Visualizing pca from the original data '''
    
    title = 'Visualizing PCA from the original images' # Title for the plot
    fig = pl.figure(figsize=(7 * n_cols, 5 * n_rows))
    fig.suptitle(title, fontweight='bold', color='red', fontsize=14)
    outer_grid = gridspec.GridSpec(5, 2)
    
    # Iterate over each class digit
    for i in range(total_class): 
        Z, U, D, M = calculate_pca(images, labels, i, x*y) # Calculate the principle components of the given class
        visualize_pca(U, D, fig, outer_grid[i], i) # Visualize PCA
        
    fig.savefig('normalpca.png') #Save the plot
    
    
    ''' Visualizing pca from noisy data '''
    
    noisy_images = [] # List to store the noisy images (moderate, very, extrem)
    denoised_images = [] # List to store the denoised images (moderate, very, extrem)
    
    for sigma in [moderately_noisy, very_noisy, extrem_noisy]:
        noisy_img = get_noisy_images(images, sigma) # Add some noise to the images
        noisy_images.append(noisy_img) # Add the noisy images to the list
        title = get_noisy_title(sigma) # Get title to be used in the plot
        fig = pl.figure(figsize=(7 * n_cols, 5 * n_rows))
        fig.suptitle(title, fontweight='bold', color='red', fontsize=14)
        outer_grid = gridspec.GridSpec(5, 2)
        
        Y = np.zeros([x, y, n]) # Matrix Y to store denoised images
        
        # Iterate over each class digit
        for i in range(total_class):
            m = get_pca_param(sigma)
            Z, U, D, M = calculate_pca(noisy_img, labels, i, m) # Calculate the principle components of the given class
            visualize_pca(U, D, fig, outer_grid[i], i) # Visualize PCA

            idx = np.where(labels[i, :] == 1)[0] # Get indices of images that belong to the given class
            im = M + np.dot(U[:, :m], Z) # Denoise images
            Y[:, :, idx] = imp.vec2im(im) # Convert 2D images to 3D images
            
        denoised_images.append(Y) # Add the denoised images to the list
        fig.savefig(get_noisy_filename(sigma)) # Save the plot
        
         
    ''' Visualize 10 examples of original images, noisy images, and denoised reconstruction '''
    
    X_moderate = noisy_images[0]
    X_very = noisy_images[1]
    X_extrem = noisy_images[2]
    
    Y_moderate = denoised_images[0]
    Y_very = denoised_images[1]
    Y_extrem = denoised_images[2]

    idx = [3, 45, 78, 105, 140, 211, 399, 480, 520, 645] # Pick 10 images to be visualized
    title = 'Visualize 10 examples of original images, noisy images, and denoised reconstruction'
    
    fig = pl.figure(figsize=(5*4, 3*2))
    fig.suptitle(title, fontweight='bold', color='red', fontsize=14)
    gs = gridspec.GridSpec(4, 2)
    ax1 = fig.add_subplot(gs[0, :])
    ax2 = fig.add_subplot(gs[1, 0])
    ax3 = fig.add_subplot(gs[1, 1])
    ax4 = fig.add_subplot(gs[2, 0])
    ax5 = fig.add_subplot(gs[2, 1])
    ax6 = fig.add_subplot(gs[3, 0])
    ax7 = fig.add_subplot(gs[3, 1])

    imp.visualize_data(ax1, images[:, :, idx], (1,10))
    ax1.set_title('Original images', fontweight='bold', fontsize=12)
    
    imp.visualize_data(ax2, X_moderate[:, :, idx], (1,10))
    ax2.set_title('Moderately noisy images', fontweight='bold', fontsize=12)
    
    imp.visualize_data(ax3, Y_moderate[:, :, idx], (1,10))
    ax3.set_title('Reconstruction moderately noisy images', fontweight='bold', fontsize=12)
    
    imp.visualize_data(ax4, X_very[:, :, idx], (1,10))
    ax4.set_title('Very noisy images', fontweight='bold', fontsize=12)
    
    imp.visualize_data(ax5, Y_very[:, :, idx], (1,10))
    ax5.set_title('Reconstruction very noisy images', fontweight='bold', fontsize=12)
    
    imp.visualize_data(ax6, X_extrem[:, :, idx], (1,10))
    ax6.set_title('Extremely noisy images', fontweight='bold', fontsize=12)
    
    imp.visualize_data(ax7, Y_extrem[:, :, idx], (1,10))
    ax7.set_title('Reconstruction extremely noisy images', fontweight='bold', fontsize=12)
    
    fig.savefig('pcaexamples.png')
    
```

```python
def calculate_pca(images, labels, cls, m):
    ''' Calculate principle components for a particular class digit '''
    
    idx = np.where(labels[cls, :] == 1)[0] # Get indices of images that belong to the given class
    images2D, fact = imp.im2vec(images[:, :, idx]) # Convert 3D images to 2D images
    Z, U, D = imp.pca(images2D, m) # Calculate principle components for the given class
    M = np.mean(images2D, axis=1)[:, np.newaxis] # Calculate the mean of the images
    
    return Z, U, D, M


def visualize_pca(U, D, fig, outer_grid, cls, k=25, n=5):
    ''' Visualize:  - all principle values 
                    - the largets 25 principle values
                    - the first k principle
    
    Parameters:
        U = (d x d) matrix of principle directions
        D = (1 x d) vector of principle values
        outer_grid = gridspec used to arrange the outer plot
        cls = class of the images
        k = number of principles values to visualize (default=25)
        n = number of principal directons to visualize (default=5)
    '''
    
    inner_grid = gridspec.GridSpecFromSubplotSpec(3, 1, subplot_spec=outer_grid, hspace=0.5)
    ax = pl.Subplot(fig, inner_grid[0])
    ax.bar(np.arange(len(D)), D)
    ax.set_title('All principle values of class ' + str(cls), fontweight='bold', fontsize=10, color='brown')
    fig.add_subplot(ax)
    
    ax = pl.Subplot(fig, inner_grid[1])
    ax.bar(np.arange(k), D[:k])
    ax.set_title('The largest 25 principle values of class ' + str(cls), fontweight='bold', fontsize=10, color='brown')
    fig.add_subplot(ax)
    
    U_3D = imp.vec2im(U[:, :n], (16, 16))
    ax = pl.Subplot(fig, inner_grid[2])
    imp.visualize_data(ax, U_3D, (1,n))
    ax.set_title('The first 5 principle directions of class ' + str(cls), fontweight='bold', fontsize=10, color='brown')
    fig.add_subplot(ax)
  
def get_noisy_images(images, sigma):
    ''' Add some noise to images with standard deviation sigma'''
    
    x, y, n = images.shape
    return images + (sigma * np.random.randn(x, y, n))    

def get_noisy_title(sigma):
    ''' Get title of the figure depending on the level of the noise'''
    
    if sigma == moderately_noisy:
        title = 'Visualizing PCA from moderately noisy images'
    elif sigma == very_noisy:
        title = 'Visualizing PCA from very noisy images'
    elif sigma == extrem_noisy:
        title = 'Visualizing PCA from extrem noisy images'
    else:
        title = 'Visualizing PCA from noisy images with standard deviation of ' + str(sigma)
    
    return title

def get_noisy_filename(sigma):
    ''' Get file name for the figure depending on the level of the noise '''
    
    if sigma == moderately_noisy:
        name = 'moderately_noisy.png'
    elif sigma == very_noisy:
        name = 'very_noisy.png'
    elif sigma == extrem_noisy:
        name = 'extremly_noisy.png'
    else:
        name = 'Unknown_noise_' + str(sigma)
    
    return name

def get_pca_param(sigma):
    ''' Get the number of the principal components 
        used as a parameter to calculate the PCA
        depending on the level of the noise 
    '''
    
    if sigma == moderately_noisy:
        m = 11
    elif sigma == very_noisy:
        m = 75
    elif sigma == extrem_noisy:
        m = 150
    else:
        m = -1
    
    return m
```

```python
def outliers_calc():
    ''' outlier analysis for assignment 5'''

    data, label = imp.load_data('banana')
    pos_idx = np.where(label == 1)[1]
    neg_idx = np.where(label == -1)[1]
    pos_data = data[:, pos_idx]
    neg_data = data[:, neg_idx]
   
    pos_data_mean = np.mean(pos_data, axis=1)
    neg_data_mean = np.mean(neg_data, axis=1)

    AUC = np.zeros([4, 3, 100])
    for rate_idx, rate in enumerate([0.01, 0.05, 0.1, 0.25]):
        print 'Calculate AUC with outliers contamination rate of ' + str(rate)
    	total_outliers = rate * pos_data.shape[1]
        for i in range(100):
            outlier_idx = np.random.choice(neg_idx, total_outliers, replace=False)
            outlier_data = data[:, outlier_idx]
            pos_data_with_outlier = np.concatenate((pos_data, outlier_data), axis=1)
            
            # First method: gamma with k = 3
            gamma_3 = imp.gammaidx(pos_data_with_outlier, 3)
            AUC[rate_idx, 0, i] = calculate_auc(gamma_3, pos_idx, outlier_idx)
            
            # Second method: gamma with k = 10
            gamma_10 = imp.gammaidx(pos_data_with_outlier, 10)
            AUC[rate_idx, 1, i] = calculate_auc(gamma_10, pos_idx, outlier_idx)
            
            # Third method: distance to the mean
            mean_data = np.mean(pos_data_with_outlier, axis=1)[:, np.newaxis]
            distance_mean = imp.distmat(mean_data, pos_data_with_outlier)
            AUC[rate_idx, 2, i] = calculate_auc(distance_mean, pos_idx, outlier_idx)
            
    np.savez_compressed('outliers', AUC=AUC)
```

```python
def calculate_auc(D, pos_idx, outlier_idx):
    min_val = np.min(D)
    max_val = np.max(D)
            
    T = np.arange(min_val, max_val, (max_val - min_val) / 100.)
           
    tpr_list = np.ones(T.shape)
    fpr_list = np.ones(T.shape)
    for idx, t in enumerate(T):
        tpr, fpr = performance_measure(D, t, pos_idx, outlier_idx)
        fpr_list[idx] = fpr
        tpr_list[idx] = tpr
                
    inds = np.argsort(fpr_list)
    fpr_list = fpr_list[inds]
    tpr_list = tpr_list[inds]
            
    auc = np.trapz(tpr_list, fpr_list)
        
    return auc

def performance_measure(X, T, actual_inlier, actual_outlier):
    total_actual_outlier = len(actual_outlier)
    total_actual_inlier = len(actual_inlier)
    
    X_outlier = X[0, total_actual_inlier:X.shape[1]]
    idx_outlier = np.where(X_outlier > T)[0]
    labels_outlier = np.zeros(total_actual_outlier)
    labels_outlier[idx_outlier] = 1
    
    X_inlier = X[0, :total_actual_inlier]
    idx_inlier = np.where(X_inlier > T)[0]
    labels_inlier = np.zeros(total_actual_inlier)
    labels_inlier[idx_inlier] = 1
    
    TP = np.sum(labels_outlier)
    FP = np.sum(labels_inlier)
    
    TPR = TP / total_actual_outlier 
    FPR = FP / total_actual_inlier
    
    return TPR, FPR
```

```python
def lle_noise():
    ''' LLE under noise for assignment 7'''
    
    data, true_emb = imp.load_data('flatroll')          # Load data set flatroll
    d, n = data.shape
    
    data_noisy_1 = data + (np.sqrt(0.2) * np.random.randn(d, n)) # data with gaussian noise 0.2
    data_noisy_2 = data + (np.sqrt(1.8) * np.random.randn(d, n)) # data with gaussian noise 1.8
    
    fig = pl.figure(figsize=(14, 18))
    
    ax1 = fig.add_subplot(321)
    ax1.scatter(data_noisy_1[0, :], data_noisy_1[1, :], s=60, c=true_emb)
    ax1.set_title('Flatroll data with variance 0.2', fontsize=14)
    
    ax2 = fig.add_subplot(322)
    ax2.scatter(data_noisy_2[0, :], data_noisy_2[1, :], s=60, c=true_emb)
    ax2.set_title('Flatroll data with variance 1.8', fontsize=14)
    
    Y_noisy_1 = imp.lle(data_noisy_1, 1, 'knn', 12, 1e-3)
    Y_noisy_2 = imp.lle(data_noisy_2, 1, 'knn', 8, 1e-3)
    
    ax3 = fig.add_subplot(323)
    ax3.scatter(true_emb, Y_noisy_1, s=60, c=true_emb)
    ax3.set_title('Resulting embedding var=0.2, k=12', fontsize=14)
    
    ax4 = fig.add_subplot(324)
    ax4.scatter(true_emb, Y_noisy_2, s=60, c=true_emb)
    ax4.set_title('Resulting embedding var=1.8, k=8', fontsize=14)
    
    Y_noisy_1_big_k = imp.lle(data_noisy_1, 1, 'knn', 50, 1e-3)
    Y_noisy_2_big_k = imp.lle(data_noisy_2, 1, 'knn', 50, 1e-3)
    
    ax5 = fig.add_subplot(325)
    ax5.scatter(true_emb, Y_noisy_1_big_k, s=60, c=true_emb)
    ax5.set_title('Resulting embedding var=0.2, k=50', fontsize=14)
    
    ax6 = fig.add_subplot(326)
    ax6.scatter(true_emb, Y_noisy_2_big_k, s=60, c=true_emb)
    ax6.set_title('Resulting embedding var=1.8, k=50', fontsize=14)
    
    fig.savefig('lle_noise.png')
```
